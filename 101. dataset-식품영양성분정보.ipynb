## 전국통합식품영양성분정보(음식)표준데이터

```python
import json
import pandas as pd
```

```python
with open('./dataset/전국통합식품영양성분정보_음식_표준데이터.json', 'r', encoding='utf-8') as file:
    data_json = json.load(file)
data_json.keys()
```

```python
data_raw = pd.DataFrame(data_json['records'])
data_raw.head()
```

```python
data_raw[data_raw['식품명'].str.startswith('삼겹살')]
```

```python
data_raw.columns, data_raw.shape
```

```python
food_names = data_raw['대표식품명'].unique().tolist()
food_names[:5],len(food_names)
```

```python
max_len = max([len(food) for food in sorted(food_names)])
print(max_len)
food_temp = []
for food in sorted(food_names):
    food_temp.append(food.rjust(max_len,' '))
    if len(food_temp) == 7:
        print('\t'.join(food_temp))
        food_temp = []
```

```python
data_raw['식품기원명'].unique()
```

```python
data = data_raw.loc[(data_raw['데이터구분코드'] == 'D') & 
                    (data_raw['식품명'] == data_raw['대표식품명']) & 
                    data_raw['식품기원코드'].str.match('2|3') &
                    (data_raw['출처코드'] != '1')]
data.head()
```
