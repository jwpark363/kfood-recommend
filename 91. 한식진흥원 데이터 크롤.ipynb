## 한식진흥원 한식 레시피 데이터
- https://www.hansik.or.kr/board/re/list/323?menuSn=193 데이터 다운로드
- 총 페이지 17, 페이지별 12개

```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.hansik.or.kr/board/re/list/323?menuSn=193&curPage='
detail_url = 'https://www.hansik.or.kr/board/re/view/323?bbsSeq='

### 상세 메뉴 URL 생성
def make_detail_url(href):
    id = href.split('(')[1][:-2]
    return detail_url + id

### 메뉴 정보 파일 저장
def save_inner_html(res,id):
    soup = BeautifulSoup(res.text)
    inner_html = str(soup.select_one('.card-body.px-4'))
    menu_name = soup.select_one('.card-body.px-4 h3').text.strip()
    with open(f'./foods/{id}_{menu_name}.text', "w", encoding="utf-8") as f:
        f.write(inner_html)
    print(f'*** create {menu_name} file')
```

```python
print('@@@ start crawling...')
for page in range(1,18):
    print(f'*** page : {page}')
    print(f'*** main : {url+str(page)}')
    res = requests.get(url+str(page))
    soup = BeautifulSoup(res.text)
    foods_url = [make_detail_url(id.attrs['href']) for id in soup.select('.stretched-link')]
    for food_url in foods_url:
        print(f'*** detail : {food_url}')
        detail_res = requests.get(food_url)
        save_inner_html(detail_res,food_url.split('=')[1])
```
