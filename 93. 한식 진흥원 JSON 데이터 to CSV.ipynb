## 한식 진흥원 JSON 데이터 To CSV
- 코드 , 메뉴, 분량, 재료(메인, 기타)

```python
import json
import os
import pandas as pd
```

```python
## read Json file
def reaed_json(file_name):
    with open(f'./json/{file_name}','r') as f:
        json_obj = json.load(f)
    json_obj['코드'] = file_name.split('_')[0]
    return json_obj
## 코드, 메뉴명, 분량, 재료 데이터 저장
def read_food_data(json_obj):
    result = { col:json_obj[col] for col in ['코드','메뉴명','분량']}
    ## 제료
    result['재료'] = ' '.join([' '.join(item['재료']) for item in json_obj['재료']])
    # for item in json_obj['재료']:
    #     if item['제목'] == '메인':
    #         print(item['재료'])
    #         result['메인재료'] = ' '.join(item['재료'])
    #     else:
    #         print(item['재료'])
    #         result['기타재료'] = ' '.join(item['재료'])
    return result
```

```python
datas = []
for file in os.listdir('./json'):
    print(file)
    obj = read_food_data(reaed_json(file))
    datas.append(obj)
df = pd.DataFrame(datas)
df.head()
```
