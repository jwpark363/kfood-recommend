# 음식 기본 정보 코드 통합 맵 작성
### 대상 : 한국음식이미지 데이터 - 한식진흥원 데이터 - 농촌진흥청 음식 기본 정보

```python
import pandas as pd
```

```python
## 한국음식이미지데이터 - 한식진흥원 데이터
kdata = pd.read_csv('./dataset/merge_df.csv')
kdata = kdata[['분류','메뉴','코드']]
kdata.columns = ['대분류','소분류','한식진흥원코드']
kdata.fillna(0,inplace=True)
kdata['한식진흥원코드'] = kdata['한식진흥원코드'].astype(int).astype(str)
kfpi_data = pd.read_csv('./dataset/kfpi_data.csv')
kfpi_data = kfpi_data[['코드','메뉴명']]
kfpi_data.columns = ['한식진흥원코드','한식진흥원메뉴']
kfpi_data['한식진흥원코드'] = kfpi_data['한식진흥원코드'].astype(str)
data1 = kdata.merge(kfpi_data,on='한식진흥원코드',how='left')
data1.fillna('',inplace=True)
data1.head()
```

```python
## 한국음식이미지데이터 - 농촌진흥청 음식 기본 정보
data2 = pd.read_csv('./dataset/food_code_map.csv')
data2 = data2[['대분류','소분류','농촌진흥원코드','농촌진흥원메뉴']]
data2.head()
```

```python
## 최종 코드 맵
merge_data = data1.merge(data2,on='소분류',how='left')
merge_data = merge_data[['대분류_x','소분류','한식진흥원코드','한식진흥원메뉴','농촌진흥원코드','농촌진흥원메뉴']]
merge_data.columns = ['대분류','소분류','한식진흥원코드','한식진흥원메뉴','농촌진흥원코드','농촌진흥원메뉴']
merge_data.head()
```

```python
code_map_json = {}
for data in merge_data.values:
    code_map_json[data[4]] = {
        'name':data[5],
        'kfood_code':data[2],
        'kfood_name':data[3],
        'image_type':data[0],
        'image_name':data[1],
        
    }
code_map_json
```

```python
import json
with open('./dataset/food_code_map.json','w') as f:
    json.dump(code_map_json,f,indent=2)
```
